```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import brier_score_loss

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)

model = MLPClassifier(random_state=42, max_iter=1000, alpha=0.001, solver="sgd", verbose=10, early_stopping=True, n_iter_no_change=10, hidden_layer_sizes=[25]).fit(X_train_sc, y_train)
p=model.predict_proba(X_test)[:,0]
brier=brier_score_loss(y_test, p)

print('Training accuracy:', model.score(X_train_sc, y_train))
print('Testing accuracy: ', model.score(X_test_sc,  y_test))

#uncertainity
print("Decision Function Shape:", model.predict_proba(X_test[:6]))
print("Brier Score:",brier)
```

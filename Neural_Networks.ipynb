```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier

cancer= load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
MLP=MLPClassifier(random_state=42, hidden_layer_sizes=[100,50],activation='tanh', solver = 'lbfgs', max_iter=10000, alpha=0.01)
MLP.fit(X_train, y_train)
print("Training Acciracy: ", MLP.score(X_train, y_train))
print("Testing Acciracy: ", MLP.score(X_test, y_test))
```

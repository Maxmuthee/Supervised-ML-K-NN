```python
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import numpy as np
cancer = load_breast_cancer()
X,y = cancer.data, cancer.target
y = np.where(y == 0,1,0)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.5)
```

```python
LR = LogisticRegression(max_iter = 10000).fit(X_train, y_train)
print("Training Score",LR.score(X_train, y_train))
print("Testing Score",LR.score(X_test, y_test))
```

```python
#RANDOM OVERSAMPLING
ROS = RandomOverSampler(random_state = 42)
X_ROS, y_ROS = ROS.fit_resample(X_train, y_train)
LR = LogisticRegression(max_iter = 10000).fit(X_ROS, y_ROS)
print("Training Score",LR.score(X_ROS, y_ROS))
print("Testing Score",LR.score(X_test, y_test))
```

```python
#SMOTE
SM = SMOTE(random_state = 42)
X_SM, y_SM = SM.fit_resample(X_train, y_train)
LR = LogisticRegression(max_iter = 10000).fit(X_SM, y_SM)
print("Training Score",LR.score(X_SM, y_SM))
print("Testing Score",LR.score(X_test, y_test))
```

```python
#RANDOM UNDERSAMPLING
RUS = RandomUnderSampler(random_state = 42)
X_RUS, y_RUS = RUS.fit_resample(X_train, y_train)
LR = LogisticRegression(max_iter = 10000).fit(X_RUS, y_RUS)
print("Training Score",LR.score(X_RUS, y_RUS))
print("Testing Score",LR.score(X_test, y_test))
```
